```python
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

# Данные
fahrenheit = [-88.6, -29.2, 32.0, 93.2, 129.2, 152.6, 212.0]
kelvins = [(f - 32) * 5 / 9 + 273.15 for f in fahrenheit]  # Преобразование в Кельвины

# Функция для построения графика
def plot_conversion(fahrenheit, kelvins):
    plt.figure(figsize=(15, 8), dpi=50)
    plt.scatter(fahrenheit, kelvins, label='Входные значения', color='blue', marker='o')
    plt.xlabel('Фаренгейт')
    plt.ylabel('Кельвины')
    plt.title('Преобразование Фаренгейта в Кельвины')
    plt.legend()
    plt.grid(True)
    plt.show()

# Построение графика
plot_conversion(fahrenheit, kelvins)

# Вывод преобразованных значений
for f, k in zip(fahrenheit, kelvins):
    print(f'Фаренгейт: {f} = Кельвины: {k}')

# Модель линейной регрессии
lr = LinearRegression()
fahrenheit_reshaped = [[f] for f in fahrenheit]
lr.fit(fahrenheit_reshaped, kelvins)

# Предсказание значений
new_fahrenheit = [[256], [123]]
predicted_kelvins = lr.predict(new_fahrenheit)

# Вывод предсказанных значений
for f, k in zip(new_fahrenheit, predicted_kelvins):
    print(f'Фаренгейт: {f[0]} => Предсказанные Кельвины: {k}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# График с областями
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, color='skyblue', alpha=0.4, label='sin(x)')
plt.fill_between(x, y2, color='sandybrown', alpha=0.5, label='cos(x)')
plt.title('График с областями')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Гистограмма случайных данных
data = np.random.randn(1000)
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Гистограмма случайных данных')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.grid(True)
plt.show()

# Диаграмма рассеяния случайных данных
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter, color='red', alpha=0.5)
plt.title('Диаграмма рассеяния случайных данных')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
import math

# 1. Вывод числа Эйлера
euler_number = math.e
print(f'Число Эйлера: {euler_number}')

# 2. Вывод числа Пи
pi_number = math.pi
print(f'Число Пи: {pi_number}')

# 3. Вывод значения NaN
nan_value = math.nan
print(f'Значение NaN: {nan_value}')

# 4. Вычисление факториала
ordinal_number = 3
factorial_value = math.factorial(ordinal_number)
print(f'Факториал числа {ordinal_number}: {factorial_value}')

# 5. Функция для нахождения наибольшего общего делителя (НОД)
def gcd(a, b):
    """Возвращает наибольший общий делитель (НОД) двух чисел."""
    while b:
        a, b = b, a % b
    return a

# Пример использования функции gcd
num1 = ordinal_number
num2 = 128
greatest_divisor = gcd(num1, num2)
print(f'Наибольший делитель для {num1} и {num2}: {greatest_divisor}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_cat_face():
    fig, ax = plt.subplots()

    # Голова
    head = plt.Circle((0.5, 0.5), 0.3, color='lightgrey', ec='black')
    ax.add_artist(head)

    # Уши
    ear_left = plt.Polygon([[0.2, 0.7], [0.3, 0.9], [0.4, 0.7]], color='lightgrey', ec='black')
    ear_right = plt.Polygon([[0.6, 0.7], [0.7, 0.9], [0.8, 0.7]], color='lightgrey', ec='black')
    ax.add_artist(ear_left)
    ax.add_artist(ear_right)

    # Глаза
    eye_left = plt.Circle((0.35, 0.55), 0.05, color='black')
    eye_right = plt.Circle((0.65, 0.55), 0.05, color='black')
    ax.add_artist(eye_left)
    ax.add_artist(eye_right)

    # Зрачки
    pupil_left = plt.Circle((0.35, 0.55), 0.02, color='white', zorder=2)
    pupil_right = plt.Circle((0.65, 0.55), 0.02, color='white', zorder=2)
    ax.add_artist(pupil_left)
    ax.add_artist(pupil_right)

    # Нос
    nose_x = [0.5, 0.48, 0.52]
    nose_y = [0.45, 0.4, 0.4]
    ax.fill(nose_x, nose_y, color='pink')

    # Улыбка
    smile_x = np.linspace(0.35, 0.65, 100)
    smile_y = -0.1 * (smile_x - 0.5)**2 + 0.38
    ax.plot(smile_x, smile_y, color='black')

    # Настройки осей
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')

    plt.title("Котик, смотрящий в душу")
    plt.show()

draw_cat_face()
```

```python
# Импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical

# Загружаем данные
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Подготовка данных
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = np.expand_dims(x_train, -1)  # Добавляем размерность канала (1)
x_test = np.expand_dims(x_test, -1)    # Добавляем размерность канала (1)
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# Создание модели
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Компиляция и обучение модели
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Оценка модели на тестовом наборе данных
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_accuracy:.4f}')

# Названия классов
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Визуализация
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(class_names[np.argmax(y_test[i])])  # Показываем название класса
    plt.axis('off')
plt.subplots_adjust(hspace=0.5)
plt.show()
```
